## Data loading

### Subtask:
Load the Tuberculosis_Trends.csv file into a Pandas DataFrame.


**Reasoning**:
Load the Tuberculosis_Trends.csv file into a pandas DataFrame and display its head and shape.



```python
import pandas as pd

try:
    df = pd.read_csv('Tuberculosis_Trends.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'Tuberculosis_Trends.csv' not found.")
    df = None
except pd.errors.ParserError:
    print("Error: Unable to parse 'Tuberculosis_Trends.csv'. Check the file format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the loaded Tuberculosis dataset to understand its characteristics.


**Reasoning**:
I need to examine the data structure, identify data types, check for missing values, and analyze key variable distributions.  This can be done in a single code block.



```python
# Display the first few rows and dimensions of the DataFrame
display(df.head())
print(df.shape)

# Print column names
print(df.columns)

# Identify data types
print(df.dtypes)

# Check for missing values
print(df.isnull().sum())
print(df.isnull().sum() / len(df) * 100)

# Analyze key variable distributions
print(df.describe())

# Distribution of TB cases over the years
display(df.groupby('Year')['TB_Cases'].describe())
display(df.groupby('Region')['TB_Cases'].describe())
# Visualizations (histograms, boxplots)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.hist(df['TB_Cases'], bins=20)
plt.xlabel('TB Cases')
plt.ylabel('Frequency')
plt.title('Distribution of TB Cases')
plt.show()


plt.figure(figsize=(12, 6))
df.boxplot(column='TB_Cases', by='Region', figsize=(12, 6))
plt.title('TB Cases by Region')
plt.suptitle('')  # Remove default title
plt.ylabel('TB Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Distribution of categorical variables
display(df['Region'].value_counts())
display(df['Income_Level'].value_counts())
```

## Data cleaning

### Subtask:
Clean the tuberculosis dataset by handling missing values and converting data types.


**Reasoning**:
Clean the data by handling missing values, converting data types, and removing duplicates.



```python
# Check for missing values
print(df.isnull().sum())

# Convert data types if necessary
for col in ['TB_Cases', 'TB_Deaths', 'TB_Incidence_Rate', 'TB_Mortality_Rate', 'TB_Treatment_Success_Rate',
            'Drug_Resistant_TB_Cases', 'HIV_CoInfected_TB_Cases', 'Population', 'GDP_Per_Capita',
            'Health_Expenditure_Per_Capita', 'Urban_Population_Percentage', 'Malnutrition_Prevalence',
            'Smoking_Prevalence', 'TB_Doctors_Per_100K', 'TB_Hospitals_Per_Million',
            'Access_To_Health_Services', 'BCG_Vaccination_Coverage', 'HIV_Testing_Coverage']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Display the first few rows and data types of the cleaned DataFrame
display(df.head())
print(df.dtypes)

# Re-check for missing values after cleaning
print(df.isnull().sum())
```

## Data wrangling

### Subtask:
Create new features and aggregate data for further analysis.


**Reasoning**:
Calculate tuberculosis rates, aggregate data by year and region, and optionally by income level if applicable.



```python
# Calculate Tuberculosis Rates (if population data is available)
if 'Population' in df.columns:
    df['TB_Incidence_Rate_Calculated'] = (df['TB_Cases'] / df['Population']) * 100000
    print("TB Incidence Rate calculated successfully.")
else:
    print("Population data not found. Skipping TB Incidence Rate calculation.")

# Aggregate Data by Year
yearly_data = df.groupby('Year').agg({'TB_Cases': 'sum'})
if 'TB_Incidence_Rate_Calculated' in df.columns:
    yearly_data['Mean_TB_Incidence_Rate'] = df.groupby('Year')['TB_Incidence_Rate_Calculated'].mean()
display(yearly_data.head())

# Aggregate Data by Region
regional_data = df.groupby('Region')['TB_Cases'].sum()
display(regional_data.head())

# Optional Aggregation by Income Level
income_level_data = df.groupby('Income_Level')['TB_Cases'].sum()
display(income_level_data)
```

## Data analysis

### Subtask:
Analyze trends in tuberculosis cases over time and across different regions/income levels.


**Reasoning**:
Analyze trends in tuberculosis cases over time using the yearly_data DataFrame. Calculate the year-over-year percentage change in TB cases. Visualize the trends with a line plot.



```python
import matplotlib.pyplot as plt

# Calculate year-over-year percentage change in TB cases
yearly_data['YOY_Change'] = yearly_data['TB_Cases'].pct_change() * 100

# Plot TB cases over time
plt.figure(figsize=(12, 6))
plt.plot(yearly_data.index, yearly_data['TB_Cases'], label='TB Cases', color='blue')
plt.plot(yearly_data.index, yearly_data['Mean_TB_Incidence_Rate'], label='Mean TB Incidence Rate', color='red')


# Plot year-over-year change
plt.twinx()  # Create a second y-axis
plt.plot(yearly_data.index, yearly_data['YOY_Change'], label='YOY Change (%)', color='green')


plt.xlabel('Year')
plt.ylabel('TB Cases')
plt.title('Trend of TB Cases Over Time')
plt.legend(loc='upper left')

plt.ylabel('YOY Change (%)', color='green')
plt.legend(loc='upper right')
plt.show()

# Analyze trends in tuberculosis cases across different regions
plt.figure(figsize=(10, 6))
regional_data.plot(kind='bar', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total TB Cases')
plt.title('Total TB Cases per Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Analyze trends in tuberculosis cases across different income levels
plt.figure(figsize=(10, 6))
income_level_data.plot(kind='bar', color='lightcoral')
plt.xlabel('Income Level')
plt.ylabel('Total TB Cases')
plt.title('Total TB Cases per Income Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Explore correlations between TB rates and other factors
correlation_matrix = df[['TB_Incidence_Rate_Calculated', 'GDP_Per_Capita', 'Health_Expenditure_Per_Capita',
                         'Malnutrition_Prevalence', 'Smoking_Prevalence', 'Access_To_Health_Services']].corr()
display(correlation_matrix)

# Scatter plots for strongest correlations
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP_Per_Capita'], df['TB_Incidence_Rate_Calculated'])
plt.xlabel('GDP Per Capita')
plt.ylabel('TB Incidence Rate')
plt.title('GDP Per Capita vs. TB Incidence Rate')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['Malnutrition_Prevalence'], df['TB_Incidence_Rate_Calculated'])
plt.xlabel('Malnutrition Prevalence')
plt.ylabel('TB Incidence Rate')
plt.title('Malnutrition Prevalence vs. TB Incidence Rate')
plt.show()
```

## Data visualization

### Subtask:
Create visualizations to illustrate the trends and insights found in the data analysis.


**Reasoning**:
Generate the visualizations as specified in the instructions, using the previously created dataframes.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Time Series Plot
plt.figure(figsize=(12, 6))
plt.plot(yearly_data.index, yearly_data['TB_Cases'], label='TB Cases', color='blue')
plt.twinx()
plt.plot(yearly_data.index, yearly_data['Mean_TB_Incidence_Rate'], label='Mean TB Incidence Rate', color='red')
plt.xlabel('Year')
plt.ylabel('TB Cases', color='blue')
plt.ylabel('Mean TB Incidence Rate', color='red')
plt.title('Trend of TB Cases and Incidence Rate Over Time')
plt.legend(loc='upper left')
plt.show()

# 2. Regional Comparison
plt.figure(figsize=(10, 6))
regional_data.plot(kind='bar', color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total TB Cases')
plt.title('Total TB Cases per Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 3. Income Level Comparison
plt.figure(figsize=(10, 6))
income_level_data.plot(kind='bar', color='lightcoral')
plt.xlabel('Income Level')
plt.ylabel('Total TB Cases')
plt.title('Total TB Cases per Income Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 4. Correlation Visualization
plt.figure(figsize=(10, 8))
sns.heatmap(df[['TB_Incidence_Rate_Calculated', 'GDP_Per_Capita', 'Health_Expenditure_Per_Capita',
                         'Malnutrition_Prevalence', 'Smoking_Prevalence', 'Access_To_Health_Services']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()


# 5. Scatter Plots for Key Correlations
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP_Per_Capita'], df['TB_Incidence_Rate_Calculated'], color='green')
plt.xlabel('GDP Per Capita')
plt.ylabel('TB Incidence Rate')
plt.title('GDP Per Capita vs. TB Incidence Rate')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['Malnutrition_Prevalence'], df['TB_Incidence_Rate_Calculated'], color='purple')
plt.xlabel('Malnutrition Prevalence')
plt.ylabel('TB Incidence Rate')
plt.title('Malnutrition Prevalence vs. TB Incidence Rate')
plt.show()
```
